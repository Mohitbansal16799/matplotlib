```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Pie chart

```python
areas = ["marketing","sales","devlopment","HR","customer"]
budget =[2.5,4,10,1,6]
plt.pie(budget,labels = areas,autopct = "%0.1if%%")
plt.show()
```

# Scatter plot

```python
x = [23,45,67,44,65,87,74,27]
y = [34,56,89,95,87,84,61,27]
plt.scatter(x,y)
plt.xlabel("age")
plt.ylabel("salary")
plt.title("age vs salary")
plt.show()
```

# Line graph

```python
product_id =[1,2,3,4,5,6,7,8]
online_price = [233,456,770,120,222,444,200,300]
offline_price = [400,300,100,333,444,521,459,889]
chor_bajar = [100,200,300,400,150,250,450,230]
```

```python
plt.plot(product_id,online_price,color = "red",marker = "+",markersize = 10,label = "online prices")
plt.plot(product_id,offline_price,color = "yellow",markersize = 10,label = "offline price")
plt.plot(product_id,chor_bajar,color = "green",marker = "D",markersize = 10,label = "chor bajar")
plt.xlabel("product id")
plt.ylabel("online price")
plt.title("comparison product id vs online price")
plt.legend()
plt.show()
```

# Bar graph

```python
year = [2012,2013,2014,2015,2016,2017,2018]
price = [65000,69000,75000,55000,52000,64000,81000]
plt.bar(year,price,color ="red")
plt.xlabel("year")
plt.ylabel("price")
plt.title("different year different price")
plt.show()
```
